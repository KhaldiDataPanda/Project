## 2- Bi-Variable Analysis

This Python script performs a series of statistical analyses on a dataset of goals and outside areas, presumably from a sports context. The data is read from a CSV file named 'goals.csv' located in a specified folder.

1. **Contingency Table**: A contingency table is created using the pandas crosstab function. This table provides a basic picture of the interrelation between the two variables, 'goals' and 'outside_areas'.

2. **Marginal Distributions**: The script calculates the marginal distributions of both 'goals' and 'outside_areas'. This gives the proportions of each variable's values.

3. **Mean and Variance**: The mean and variance for both 'goals' and 'outside_areas' are calculated. These are basic statistical measures that provide information about the dataset's central tendency and dispersion.

4. **Correlation Coefficient**: The script calculates the correlation coefficient between 'goals' and 'outside_areas'. This value indicates the strength and direction of the linear relationship between the two variables.

5. **Linear Regression**: A linear regression is performed on 'goals' and 'outside_areas'. The output includes the slope and intercept of the regression line, the correlation coefficient (r_value), the p-value for a hypothesis test whose null hypothesis is that the slope is zero, and the standard error of the estimated gradient.

6. **T-Test**: A t-test is performed to compare the means of 'goals' and 'outside_areas'. The output is the t-statistic and the p-value.

7. **ANOVA (Analysis of Variance)**: An ANOVA test is performed on 'goals' and 'outside_areas'. The output is the F-statistic and the p-value. This test is used to determine whether there are any statistically significant differences between the means of two or more groups.

8. **Chi-Square Test of Independence**: A Chi-Square test of independence is performed on the contingency table. The output is the Chi2-statistic and the p-value. This test is used to determine if there is a significant association between two categorical variables.

Finally, the script prints all the results.

```python
import pandas as pd
import scipy.stats as stats

# Load the goals dataframe
goals_df = pd.read_csv('goals.csv')

# Create a contingency table
contingency_table = pd.crosstab(goals_df['right_foot'], goals_df['left_foot'])

# Calculate marginal distributions
marginal_right_foot = goals_df['right_foot'].value_counts(normalize=True)
marginal_left_foot = goals_df['left_foot'].value_counts(normalize=True)

# Calculate mean and variance for each variable
mean_right_foot = goals_df['right_foot'].mean()
mean_left_foot = goals_df['left_foot'].mean()
var_right_foot = goals_df['right_foot'].var()
var_left_foot = goals_df['left_foot'].var()

# Calculate correlation coefficient
correlation_coefficient = goals_df['right_foot'].corr(goals_df['left_foot'])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(goals_df['right_foot'], goals_df['left_foot'])

# Perform t-test
t_statistic, p_value_ttest = stats.ttest_ind(goals_df['right_foot'], goals_df['left_foot'])

# Perform ANOVA
f_statistic, p_value_anova = stats.f_oneway(goals_df['right_foot'], goals_df['left_foot'])

# Perform chi-square test of independence
chi2_statistic, p_value_chi2, _, _ = stats.chi2_contingency(contingency_table)

# Print the results
print("Contingency Table:")
print(contingency_table)
print("\nMarginal Distributions:")
print("Right Foot:", marginal_right_foot)
print("Left Foot:", marginal_left_foot)
print("\nMean and Variance:")
print("Right Foot - Mean:", mean_right_foot)
print("Right Foot - Variance:", var_right_foot)
print("Left Foot - Mean:", mean_left_foot)
print("Left Foot - Variance:", var_left_foot)
print("\nCorrelation Coefficient:", correlation_coefficient)
print("\nLinear Regression:")
print("Slope:", slope)
print("Intercept:", intercept)
print("R-value:", r_value)
print("P-value:", p_value)
print("Standard Error:", std_err)
print("\nT-test:")
print("T-statistic:", t_statistic)
print("P-value:", p_value_ttest)
print("\nANOVA:")
print("F-statistic:", f_statistic)
print("P-value:", p_value_anova)
print("\nChi-square Test of Independence:")
print("Chi2-statistic:", chi2_statistic)
print("P-value:", p_value_chi2)
```

## 3- Multi-Variable Analysis

This Python script is performing Principal Component Analysis (PCA) on a dataset of goals in football (soccer). PCA is a technique used in exploratory data analysis and for making predictive models. It is a statistical procedure that orthogonally transforms the 'n' coordinates of a dataset into a new set of 'n' coordinates known as the Principal Components.

1. **Importing necessary libraries**: The script begins by importing the necessary Python libraries - pandas for data manipulation, numpy for numerical operations, PCA from sklearn.decomposition for performing PCA, and matplotlib.pyplot for data visualization.

2. **Loading the dataset**: The 'goals.csv' dataset is loaded into a pandas DataFrame.

3. **Data preprocessing**: The relevant columns for the analysis are selected. The data is then centered and scaled. Centering is subtracting the mean of the column from each value, and scaling is dividing by the standard deviation. This is done to normalize the data.

4. **Covariance and Correlation Matrix Calculation**: The covariance and correlation matrices of the scaled data are calculated. These matrices provide insights into the relationships between different variables in the dataset.

5. **Eigenvalue Decomposition**: The eigenvalues and eigenvectors of the covariance matrix are calculated. These are used to identify the principal components of the data.

6. **Sorting Eigenvalues and Eigenvectors**: The eigenvalues and eigenvectors are sorted in descending order. The eigenvalues indicate the amount of variance in the data that is associated with each principal component, and the eigenvectors define the direction of the principal components.

7. **Selection of Factorial Axes**: The explained variance ratio and cumulative variance ratio are calculated. These ratios provide insights into how much of the total variance in the data is explained by each principal component.

8. **Elbow Method**: A scree plot is created to visualize the explained variance ratio for each principal component. This helps in determining the optimal number of principal components to retain.

9. **Explained Variance Criterion and Kaiser's Method**: Two methods are used to determine the number of principal components to retain - the explained variance criterion (retain components until 95% of the variance is explained) and Kaiser's method (retain components with eigenvalues greater than 1).

10. **Performing PCA**: PCA is performed on the scaled data using the selected number of components.

11. **Visualization**: Scatter plots (Individuals' cloud and Correlation circle) are created to visualize the principal components and the correlation between the original variables and the principal components.

12. **Quality of Representation and Contribution of an Individual**: The quality of representation of the individual and the contribution of an individual are calculated. These provide insights into how well each individual (data point) is represented by the principal components and how much each individual contributes to the principal components.

13. **Correlation between Original Variables and Axes**: The correlation between the original variables and the axes is calculated. This provides insights into how each original variable is related to the principal components.

14. **Printing the Results**: Finally, the results are printed to the console for further analysis.

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
# Load the goals dataset
goals_df = pd.read_csv('goals.csv')

# Select the relevant columns for analysis
selected_columns = ['right_foot', 'left_foot', 'inside_area', 'outside_areas','penalties']

# Centering and scaling the data
scaled_data = (goals_df[selected_columns] - goals_df[selected_columns].mean()) / goals_df[selected_columns].std()

# Calculation of the covariance matrix
covariance_matrix = scaled_data.cov()

# Calculation of the correlation matrix
correlation_matrix = scaled_data.corr()

# Eigenvalue decomposition
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# Selection of factorial axes
explained_variance_ratio = sorted_eigenvalues / np.sum(sorted_eigenvalues)
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# Elbow method to determine the optimal number of principal components
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()

# Explained variance criterion
desired_variance_ratio = 0.95
num_components = np.argmax(cumulative_variance_ratio >= desired_variance_ratio) + 1

# Kaiser's method to select components with eigenvalues greater than 1
kaiser_components = np.sum(sorted_eigenvalues > 1)

# Perform PCA with the selected number of components
pca = PCA(n_components=num_components)
principal_components = pca.fit_transform(scaled_data)

# Individuals' cloud (scatter plot of principal components)
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Individuals\' Cloud')
plt.show()

# Correlation circle (scatter plot of original variables and principal components)
plt.scatter(pca.components_[0], pca.components_[1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Correlation Circle')
plt.show()

# Quality of representation of the individual
individual_quality = np.sum(principal_components ** 2, axis=1)

# Contribution of an individual
individual_contribution = individual_quality / np.sum(individual_quality)

# Correlation between original variables and axes
variable_correlation = np.abs(np.dot(scaled_data, pca.components_.T))

# Print the results
print("Centered and Scaled Data:")
print(scaled_data.head())
print("\nCovariance Matrix:")
print(covariance_matrix)
print("\nCorrelation Matrix:")
print(correlation_matrix)
print("\nEigenvalues:")
print(sorted_eigenvalues)
print("\nExplained Variance Ratio:")
print(explained_variance_ratio)
print("\nCumulative Variance Ratio:")
print(cumulative_variance_ratio)
print("\nNumber of Principal Components (Explained Variance >= 0.95):", num_components)
print("\nNumber of Principal Components (Kaiser's Method):", kaiser_components)
print("\nPrincipal Components:")
print(principal_components)
print("\nIndividual Quality:")
print(individual_quality)
print("\nIndividual Contribution:")
print(individual_contribution)
print("\nVariable Correlation:")
print(variable_correlation)
```
